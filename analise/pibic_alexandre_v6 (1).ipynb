```python
import numpy as np
from google.colab import autoviz
df_5972375734822291333 = autoviz.get_df('df_5972375734822291333')

def time_series_multiline(df, timelike_colname, value_colname, series_colname, figsize=(2.5, 1.3), mpl_palette_name='Dark2'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette(mpl_palette_name))
  def _plot_series(series, series_name, series_index=0):
    if value_colname == 'count()':
      counted = (series[timelike_colname]
                 .value_counts()
                 .reset_index(name='counts')
                 .rename({'index': timelike_colname}, axis=1)
                 .sort_values(timelike_colname, ascending=True))
      xs = counted[timelike_colname]
      ys = counted['counts']
    else:
      xs = series[timelike_colname]
      ys = series[value_colname]
    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

  fig, ax = plt.subplots(figsize=figsize, layout='constrained')
  df = df.sort_values(timelike_colname, ascending=True)
  if series_colname:
    for i, (series_name, series) in enumerate(df.groupby(series_colname)):
      _plot_series(series, series_name, i)
    fig.legend(title=series_colname, bbox_to_anchor=(1, 1), loc='upper left')
  else:
    _plot_series(df, '')
  sns.despine(fig=fig, ax=ax)
  plt.xlabel(timelike_colname)
  plt.ylabel(value_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = time_series_multiline(df_5972375734822291333, *['last_update', 'count()', 'update_status'], **{})
chart
```

```python
import numpy as np
from google.colab import autoviz
df_1408804363314599985 = autoviz.get_df('df_1408804363314599985')

def categorical_histogram(df, colname, figsize=(2, 1.2), mpl_palette_name='Dark2'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  _, ax = plt.subplots(figsize=figsize)
  bars = df[colname].value_counts()
  for bar, value in zip(bars.index, bars.values):
    percentage = round(100 * value / len(df), 2)
    label = f'{value} ({percentage}%)'
    plt.annotate(label, (bar, value), ha='center', va='bottom')
  plt.barh(bars.index, bars.values, color=sns.palettes.mpl_palette(mpl_palette_name))
  plt.title(colname)
  ax.spines[['top', 'right',]].set_visible(False)
  return autoviz.MplChart.from_current_mpl_state()

chart = categorical_histogram(df_1408804363314599985, *['update_status'], **{})
chart
```

```python
import os
from PIL import Image

def png_to_pdf(input_file, output_file):
  """Converte um arquivo PNG para PDF.

  Args:
    input_file: O caminho do arquivo PNG de entrada.
    output_file: O caminho do arquivo PDF de saída.

  Returns:
    O caminho do arquivo PDF de saída.
  """

  # Abre o arquivo PNG de entrada.
  img = Image.open(input_file)

  # Converte o arquivo PNG para PDF.
  img.save(output_file, 'pdf')

  return output_file


if __name__ == '__main__':
  # Obtém o caminho do arquivo PNG de entrada.
  input_file = input('Digite o caminho do arquivo PNG de entrada: ')

  # Obtém o caminho do arquivo PDF de saída.
  output_file = input('Digite o caminho do arquivo PDF de saída: ')

  # Converte o arquivo PNG para PDF.
  output_file = png_to_pdf(input_file, output_file)

  # Imprime o caminho do arquivo PDF de saída.
  print('O arquivo PDF foi criado em:', output_file)
```
