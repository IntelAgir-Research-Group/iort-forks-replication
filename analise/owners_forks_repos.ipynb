```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
file_path = 'csv/owners_qtde_repos.csv'
data = pd.read_csv(file_path)
```

```python
# Gerar o gráfico de distribuição
plt.figure(figsize=(10, 6))
plt.hist(data['public_repos'], bins=1000, color='skyblue', edgecolor='black')
plt.xlabel('Quantidade de Repositórios Públicos', fontsize=14)  # Aumentar o tamanho da fonte do rótulo x
plt.ylabel('Número de Donos de Forks', fontsize=14)  # Aumentar o tamanho da fonte do rótulo y
plt.xlim([0, 2000])

plt.grid(True)

plt.savefig('figs/owners_repos.pdf', format='pdf')

plt.show()
```
