```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o CSV para um DataFrame
file_path = "csv/forks_atualizacoes_disponiveis.csv"  # Substitua pelo caminho real do seu arquivo CSV
df = pd.read_csv(file_path)
```

```python
# Converter a coluna "commits_behind" para numérico
df["commits_behind"] = pd.to_numeric(df["commits_behind"], errors="coerce")

# Análise estatística
stats = df.groupby("disponivel")["commits_behind"].describe()
```

```python
# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="disponivel", y="commits_behind")
plt.title("Relação Estatística entre commits atrasados e disponibilidade dos forks.")
plt.ylabel("Commits Atrasados")
plt.xlabel("Disponível")
plt.tight_layout()

plt.savefig('figs/atrasados_disponiveis.pdf', format='pdf')

plt.show()

#print(stats)
```
